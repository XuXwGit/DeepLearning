```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
def get_position_encoding(max_seq_len, emb_dim):
    position_encoding = np.array([[pos / pow(10000, 2 * i/emb_dim) for i in range(emb_dim)] for pos in range(max_seq_len)])
    position_encoding[1:, 0::2] = np.sin(position_encoding[1:, 0::2])
    position_encoding[2:, 1::2] = np.cos(position_encoding[2:, 1::2])

    return position_encoding
```

```python
position_encoding = get_position_encoding(100, 10)
plt.figure(figsize=(10, 10))
sns.heatmap(position_encoding)
```
